판다스를 이용한 전처리

```python
# 파일 열기
f = open('data.txt', 'r')  # r은 읽기 모드.. 파일이 없으면 오류
f = open('data.txt', 'w')  # w은 쓰기 모드.. 파일이 없으면 새로 생성, 있으면 기존내용을 덮어쓴다
f = open('data.txt', 'rt+')  # 읽기 쓰기모드로 기존파일의 맨 뒤에서부터 내용을 추가
f = open('data.txt', 'wt+')  # 읽기 쓰기모드로 기존파일의 내용을 지우고 처음부터 쓴다
f.write('hello')  # 파일에 기록
f.close()  # 파일을 닫는다
```

### 람다 함수
  - 익명의 함수
  - 함수이름은 없고 함수의 동작만 정의
  - 함수 자체를 매개변수로 하는 또 다른 함수에서 사용
  - 불필요한 함수를 미리 만들지 않아도.. 사용하고 없어진다.

```python
def myfunc(x):
  return x+1
```

```python
list_1 = [1,3,5,7,9]
list( map(myfunc,list_1) )
```

```python
list( map( lambda x: x+1 ,list_1) )
```

```python
def filter_5_over(x):
  return x > 5

list( filter(filter_5_over, list_1) )
```

```python
list( filter(lambda x: x>5, list_1) )
```
