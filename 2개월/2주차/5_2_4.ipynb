판다스를 이용한 전처리

```python
# 파일 열기
f = open('data.txt', 'r')  # r은 읽기 모드.. 파일이 없으면 오류
f = open('data.txt', 'w')  # w은 쓰기 모드.. 파일이 없으면 새로 생성, 있으면 기존내용을 덮어쓴다
f = open('data.txt', 'rt+')  # 읽기 쓰기모드로 기존파일의 맨 뒤에서부터 내용을 추가
f = open('data.txt', 'wt+')  # 읽기 쓰기모드로 기존파일의 내용을 지우고 처음부터 쓴다
f.write('hello')  # 파일에 기록
f.close()  # 파일을 닫는다
```
