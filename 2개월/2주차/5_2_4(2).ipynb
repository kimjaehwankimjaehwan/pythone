```python
import pandas as pd
```

```python
df = pd.read_csv('https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv')
df.head()
```

```python
# 결측치 확인
df.info()
```

```python
# 결측치 시각화  컬럼이 많고 데이터도 많을때 결측치를 시각화
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
import numpy as np
df['test'] = np.nan
df.loc[:200,'test'] = 0
msno.matrix(df)
plt.show()
```

```python
plt.figure(figsize=(13,5))
missing_data = df.isna().mean()
sns.barplot(x=missing_data.index, y=missing_data)
plt.xticks(rotation=45)
plt.show()
```

```python
# 기본 통계분석

print('결측치 데이터타입')
print(df.info())
print('수치형 데이터의 통계요약')
print(df.describe())
print('범주형 데이터의 통계요약')
print(df.describe(include='object'))
```
