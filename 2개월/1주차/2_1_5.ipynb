# 시각화

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = list(range(20))
y = np.random.randint(1,100,20)
print(x)
print(y)
# 선 그래프
plt.figure(figsize=(15,5))
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Graph')
plt.show()
```

```python
# 막대 그래프
categorys = ['A','B','C','D','E']
values = np.random.randint(1,100,5)
plt.figure(figsize=(10,5))  # 크기설정
plt.bar(categorys,values) # 그래프 종류
plt.xlabel('Category') # x축 라벨
plt.ylabel('Value') # y축 라벨
plt.title('Bar Graph') # title
plt.show() # 객체정보 삭제
```

```python
# 산점도 - 클러스터링 즉 군집을 시각화 할때
x = np.random.randint(1,100,50)
y = np.random.randint(1,100,50)
plt.figure(figsize=(10,5))  # x y 축 정보를 셋팅 - figsize는 가로세로의 크기를 지정
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.show()
```

```python
# 히스토그램 - 계급구간  즉. 구간을 나눠서 시각화
data = np.random.randn(1000)  # 표준 정규분포를 따르는 1000개의 난수를 생성하는 함수  평균이 0 이고 표준편차가 1인

plt.figure(figsize=(10,5))

plt.hist(data,bins=20)  # 정규분포에서 계급구간은 20개로 설정
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
# 파이 차트
plt.figure(figsize=(10,5))
labels = ['A','B','C','D']
size = [15,30,45,10]
colors = ['red','orange','yellow','green']
plt.pie(size,labels=labels,colors=colors,autopct='%.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
# 박스 플롯  4분위수로 나눠서 시각화(데이터의 분포)  이상치를 판단하는 근거
# 1 2 3 4 100
data = np.random.normal(size=100)

plt.figure(figsize=(10,5))

plt.boxplot(data)
plt.title('Box Plot')
plt.show()
```

```python
# 히트맵  -  상관관계 지수를 시각활때
data = np.random.rand(10,10)
plt.figure(figsize=(10,5))
plt.imshow(data,cmap='hot',interpolation='nearest')
plt.colorbar()
plt.show()
```

```python
# 실습
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
url = 'https://query1.finance.yahoo.com/v7/finance/download/AAPL?period1=1595721600&period2=1627257600&interval=1d&events=history'
data = pd.read_csv(url)
data.head()
```

```python
data.info()
```

```python
plt.figure(figsize=(15,5))
# 문자열 날자를 날자데이터로 변경
data['Date'] = pd.to_datetime(data['Date'])

plt.plot(data['Date'],data['Close'])
plt.xlabel('Date')
plt.ylabel('Close')
plt.xticks(rotation=45)  # x축의 값을 90도 회전
plt.title('AAPL Stock Price')
plt.show()
```
