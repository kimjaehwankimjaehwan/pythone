```python
# 학생클래스
# 클래스명 Student
# 필드 : name, age , score
# 메서드 : getTotalScore, getAverage
class Student:
  name = ''
  age = 0
  score = []
  def getTotalScore(self):  # 객체로 호출할때는 객체가 자신이므로 self로 전달
    return sum(self.score)
  def getAverage(self):
    return self.getTotalScore() /len(self.score)  # 클래스 안에서 자신의 메소드를 호출할때는 자동으로 전달
```

```python
# 클래스 내부에서 메소드를 정의할때 첫번재는 반드시 self
# 사용할때는 self를 고려하지 않는다
# ex)  def myMehod(self)  # 매개변수가 없는 메소드
#  object.myMethod()
```

```python
st1 = Student()
st1.name = '홍길동'
st1.age = 20
st1.score = [85,95,88,89]
print(st1.getTotalScore())  # 메소드를 사용할때 첫번째 매개변수는 st1 전달된다
print(st1.getAverage())
```

```python
# 커피자판기
# 돈을넣고 버튼을 누르면 3개중에 한개가 나온다

# 클래스명 : CoffeMachine
# 필드  : money
# 메소드 : def choiceCoffe   리턴값 : 블랙,설탕,믹스
```

```python
class CoffeMachine:
  money = 0  # 자판기에 있는 돈
  def choiceCoffe(self,choiceNum):
    if choiceNum == 1:
      self.money += 100
      return '블랙'
    elif choiceNum == 2:
      self.money += 200
      return '설탕'
    elif choiceNum == 3:
      self.money += 300
      return '믹스'
```

```python
c = CoffeMachine()
c.choiceCoffe(3)
```

```python
# 객체지향
# 학생 한명에 대한 클래스  Student
# 이름,나이,성별,주소,과목별 점수(국 영 수)

# 학생들 관리하는 클래스 StudentMng
# 학생들 리스트, 최고점수를 받은 학생을 리턴하는 메소드, 전체학생들의 평균
```

```python
# 클래스의 필드중에서 주소기반의 변수는 모든 클래스가 공유한다.
# 단독으로 사용하려면 메소드 안에서 self.변수 형태로 정의
class Student:
  name = ''
  age = 0
  gender = ''
  address = ''
  def setScore(self,kor,eng,math):
    self.score = {}             # 지역변수가 아니라.. 이변수는 setScore가 실행되면 생성되는 변수
    self.score['kor'] = kor
    self.score['eng'] = eng
    self.score['math'] = math

  def getTotalScore(self):
    return sum(self.score.values())
  def showStudent(self):
    print(f"이름:{self.name} 나이:{self.age} 성별:{self.gender} 주소:{self.address} 점수:{self.score}")

class StudentMng:
  students = []
  def findTop1(self):  # 성적이 가장 높은 학생 한명을 찾는 함수
    maxScore = 0  # 최대값을 저장하는 변수
    maxStudent = None  # 최대값을 가진 학생을 저장하는 변수
    for student in self.students:   # 학생객체들을 순환
      if maxScore < student.getTotalScore():
        maxScore = student.getTotalScore()
        maxStudent = student
    return maxStudent, maxScore

  def getAverageStudents(self):
    return sum(student.getTotalScore() for student in self.students) / len(self.students)
```

```python
# 학생 1명
s1 = Student()
s1.name = '홍길동1'
s1.age = 20
s1.gender = '남'
s1.address = '서울'
s1.setScore(83,93,88)
# 학생 2명
s2 = Student()
s2.name = '홍길동2'
s2.age = 22
s2.gender = '여'
s2.address = '부산'
s2.setScore(75,85,88)
# 학생 3명
s3 = Student()
s3.name = '홍길동3'
s3.age = 23
s3.gender = '남'
s3.address = '경기'
s3.setScore(85,92,100)
# 관리객체
sm = StudentMng()
sm.students = [s1,s2,s3]
```

```python
# 모든 학생의 정보를 출력
for s in sm.students:
  s.showStudent()
# 최고 성적을 받은 학생의 정보를 출력
student,score = sm.findTop1()

student.showStudent()
# 전체 학생의 평균
sm.getAverageStudents()
```
