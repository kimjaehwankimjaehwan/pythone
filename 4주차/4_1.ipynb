- 객체지향 : Object 단위로 프로그램
  - 클래스 : Object를 만드는 기본적인 단위

```python
student_score = [80,70,60,90]
student_name = ['a','b','c','d']

def find_max_score(student_score):
  max_score = max(student_score)
  max_index = student_score.index(90)
  return student_name[max_index], max_score
```

```python
# 설계도..
class Student:
  student_score = [80,70,60,90]
  student_name = ['a','b','c','d']

  def find_max_score(self):
    max_score = max(student_score)
    max_index = student_score.index(90)
    return student_name[max_index], max_score
```

```python
# 클래스를 이용해서 객체를 생성하고
# 생성된 객체를 이용해서 프로그래밍
student1 = Student()
student2 = Student()
student3 = Student()
```

```python
student1.find_max_score()
```

```python
# 자동차
# 색상,현재속도  --> 상태를 나타내는 변수
# 값을 받아서 속도를증가, 속도를 감소  method
class Car:
  color = ""
  speed = 0
  def upSpeed(self, value):
    self.speed += value
  def downSpeed(self,value):
    self.speed -= value
```

```python
# Car 클래스의 객체를 생성
c1 = Car()
```

```python
# 객체를 통해서 메소드를 실행하고 변수에 적용
c1.upSpeed(10); c1.upSpeed(10); c1.upSpeed(10)
```

```python
# 객체의 변수값을 확인
c1.speed
```
