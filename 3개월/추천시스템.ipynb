```python
# TF-IDF 알고리즘 : 특정 문서에서 단어의 중요도
# TF(Term Frequency, 단어빈도)  : 특정문서에서 특정 단어가 얼마나 자주 등장하는지를 나타낸다
# TF = 특정단어 T의 빈도 / 문서 D의 총 단어수    3/100 = 0.03
# IDF(Inverse Document Frequency, 역문서 빈도) : 단어가 여러문서에서 얼마나 드물게 등장하는지를 측정
# IDF(t) = log(N/df(t))
```

```python
docs = [
    '먹고 싶은 사과',
    '먹고 싶은 바나나',
    '길고 노란 바나나 바나나',
    '저는 과일이 좋아요'
]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer()
```

```python
countvect = vect.fit_transform(docs)
countvect
```

```python
countvect.toarray()
```

```python
vect.vocabulary_
```

```python
sorted(vect.vocabulary_)
```

```python
import pandas as pd
countvect_df = pd.DataFrame(countvect.toarray(),columns=sorted(vect.vocabulary_))
countvect_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(countvect_df,countvect_df)
```

```python
# TF-IDF 수행
from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer()
tfvect = vect.fit(docs)
```

```python
tfidv_df =  pd.DataFrame( tfvect.transform(docs).toarray(), columns=sorted(vect.vocabulary_) )
tfidv_df.index = ['문서1','문서2','문서3','문서4']
tfidv_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(tfidv_df,tfidv_df)
```

```python
vect = TfidfVectorizer(max_features=4)
tfvect = vect.fit(docs)
```

```python
tfidv_df = pd.DataFrame(tfvect.transform(docs).toarray(),columns=sorted(vect.vocabulary_))
tfidv_df.index = ['문서1','문서2','문서3','문서4']
tfidv_df
```

# Word2Vec 알고리즘

```python
from gensim.models import Word2Vec
```

```python
sentence = [
    ["i","love","machine","learning"],
    ["i","enjoy","deep","learning"],
    ["i","love","coding","in","Python"],
]
# Word2Vec 학습
model = Word2Vec(sentence,vector_size=100, window=5,min_count=1,workers=4)

# 단어 'love'의 벡터 값
vector_love = model.wv['love']
# machine 과 learning의 유사도 값 계산
similarity = model.wv.similarity('machine','learning')
# 가장 유사한 단어 찾기
model.wv.most_similar('machine')
```

### 영화 추천시스템 Word2Vec

```python
path = '/content/ratings.csv'
movie = pd.read_csv(path,low_memory=False)
movie.head()
```

```python
movie.sort_values(by='timestamp',inplace=True)
movie = movie.reset_index(drop=True)
movie.head()
```

```python
pd.options.display.max_columns = None
```

```python
# 영화 제목을 가져오기 위해서...
meta = pd.read_csv('/content/movies_metadata.csv',low_memory=False)
meta.head(2)
```

```python
movie['movieId'] = movie['movieId'].astype(str)
meta['id'] = meta['id'].astype(str)
```

```python
movie.shape, meta.shape
```

```python
# movie2 = movie.iloc[:50000]
# movie2.shape
```

```python
movie = pd.merge(movie,meta[['id','original_title']],left_on='movieId',right_on='id',how='left')
movie.head(2)
```

```python
movie.drop(columns=['id'],inplace=True)
movie.head(2)
```

```python
movie = movie.dropna(subset=['original_title'])
agg = movie.groupby(['userId'])['original_title'].agg({'unique'})
agg.head()
```

```python
# Word2Vec 적용 : int 타입은 학습이 안되요... str 변경
sentence = []
for user_sentence in agg['unique'].values:
  sentence.append(list(map(str,user_sentence)))
# word2vec 의 학습을 진행
from gensim.models import Word2Vec
embedding_model = Word2Vec(sentence,window=5,vector_size=20, min_count=1,workers=4)

```

```python
sentence
```

```python
embedding_model.wv.most_similar('Madagascar')
```
