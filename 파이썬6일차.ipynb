```python
def coffee_machine(button):
  print("#1 뜨거운 물을 준비한다")
  print("#2 종이컵을 준비한다")

  if button ==1 :
    print("#3.보통 커피를 탄다")
  elif button ==2 :
    print("#3.설탕 커피를 탄다")
  elif button ==3 :
    print("#3.블랙 커피를 탄다")
  else:
    print("#3.아무거나 탄다")

  print("#4.물을 붇는다")
  print("#5.스푼으로 젓는다")
  print()
  print("#6.손님에게 가져다 준다")
```

```python
# 함수
# mycalc2
# argument / parameter   정수 두개와 연사자기호(+ - * /)  3개
# 사용  mycalc2(3,5,'/')
```

```python
def mycalc2(num1,num2, opt ):
  if opt == '+':
    return num1 + num2
  elif opt == '-':
    return num1 - num2
  elif opt == '*':
    return num1 * num2
  elif opt == '/' and  num2 == 0:
      print("0으로 나눌수 없습니다.")
      return None
  elif opt == '/':
    return num1 / num2
  else:
    print("+ - * / 만 입력하세요")
```

```python
mycalc2(10,2,'/')
```

```python
# 주어진 리스트에서 가장 큰 수를 찾는 기능
import random
random_list = [random.randint(1,100) for i in range(5)]
print(random_list)

max_num = random_list[0]
for i in random_list:
  if max_num < i:
    max_num = i

print(max_num)
```

```python
def find_max( lists ):
  max_num = lists[0]
  for i in lists:
    if max_num < i:
      max_num = i
  return max_num

# 사용

import random
datas = [random.randint(1,100) for i in range(5)]
print(datas)
max_number = find_max(datas)
print(max_number)
```
