## stack의 적용분야(예제)

```python
# {([({})])}
# 괄호가 짝이 맞는지 조사
stack = []
express = '((())})'
def is_balance(express):
  for i in express:
    if i in '[{(':  # 여는 괄호면 스텍에 쌓는다
      push(i)
    elif i in ')}]':
      p = pop()
      if ( i == ')' and p != '(' ) or ( i=='}' and p != '{' ):
        return False
  return isEmpty()

is_balance(express)
```
