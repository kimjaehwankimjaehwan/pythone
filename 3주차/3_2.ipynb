```python
temp = []  # 전역변수  global 변수
def temp_access(data):
  temp.append(data)
print(temp)
```

```python
# temp 리스트를 초기화하고 temp_access함수를 통해서 temp리스트에 데이터를 추가
temp = [] # 전역변수 temp 초기화
def append_data():
  # temp = []
  for i in range(10):
    temp_access(i)     # temp리스트에 0 ~ 9 까지 10개의 데이터가 저장

  print(f"append_data 안에서 출력한 temp의 값 : {temp}")

append_data()
print(f"전역변수 temp : {temp}")
```

# 교집합, 합집합, 차집합

```python
# 두개의 리스트 A, B가 있다. A에있는 데이터중에 B에 있는 데이가 있으면 A에서 제외해라
# 예 A = [1,2,3,4,5]  B=[3,5,7,9,10]   =>  [1,2,4]
# A - B 차집합을 구해보자
A = [1,2,3,4,5] ;  B=[3,5,7,9,10]
```

```python
# Set 데이터 들사이에서는
# 교집합 &   intersection
# 합집합 |   union
# 차집합 -   difference
# 대칭차집합 ^   symetric_difference
```

```python
set_a = set(A);  set_b = set(B)
print(set_a & set_b)
print(set_a | set_b)
print(set_a - set_b)
print(set_a ^ set_b)
```

```python
print(set_a.intersection(set_b) )
print(set_a.union( set_b ))
print(set_a.difference( set_b) )
print(set_a.symmetric_difference( set_b))
```

```python
subject_list = ['국어','영어','수학']
score_lilst = [100,85,95]

my_dic = {}
for i in range(3):
  key = subject_list[i]
  value = score_lilst[i]
  my_dic[key] = value

my_dic
```

```python
# 학생성적 리스트
name = ['김','이','박']
score = [
    [100,50,90],[90,80,90],[80,90,50]
]
# key : name, score
# value : ['김','이','박'], [   [100,50,90],[90,80,90],[80,90,90] ]

# students = [
#  {'name':'김', 'score':[100,50,90]},
#   {'name':'이', 'score':[90,80,90]},
#   {'name':'박', 'score':[80,90,90]}
# ]

students = []
for idx,name in enumerate(name):
  student = {}
  student['name'] = name
  student['score'] = score[idx]
  student['mean'] = round( sum(score[idx]) / len(score[idx]), 2)
  students.append(student)
students

```

```python
# 각 학생의 평균을 구해서 새로운 key:value형태로  'mean':000
# 평균이 가장 높은 학생을 찾아라
```
